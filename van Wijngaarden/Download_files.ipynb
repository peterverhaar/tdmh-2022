```python
import requests
from bs4 import BeautifulSoup
import re

def download_texts( title , url ):
    print( f'Downloading {title} ...')
    title = re.sub( r'\s+' , '_' , title )
    out = open( f'{title}.txt' , 'w' , encoding = 'utf-8' )
    main_url = url

    response = requests.get(main_url)
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup( html ,"lxml")

        main_text = soup.find('div', {'class': 'contentholder'} )

        #print(main_text)
        links = main_text.find_all( 'a' )

        for l in links:
            relative_url = l.get('href')
            #print(url.text , url.get('href'))
            if re.search( r'php$', relative_url ):
                print(main_url + relative_url) 
                response2 = requests.get( main_url + relative_url)
                if response2.status_code == 200:
                    html2 = response2.text
                    #print(html)
                    soup2 = BeautifulSoup( html2 ,"lxml")
                    main_text = soup2.find('table', {'id': 'tekst_en_noten'} )
                    full_text = main_text.text.strip()
                    lines = re.split( r'\n' , full_text)
                    full_text = ''
                    for i,l in enumerate(lines):
                        if re.search( r'^(voetnoot\*)' , l ):
                            break
                        if re.search( r'^(margenoot\+)' , l ):
                            l = ''
                        if re.search( r'\w+' , l ):
                            l = re.sub( r'\[pagina\s+\*?\w+\]' , '', l )
                            l = re.sub( r'\[pagina\s+(\*)*\w+\]' , '', l )
                            l = re.sub( r'\[p.\s+(\*)*\w+\]' , '', l )
                            l = re.sub( r'Ga naar voetnoot\*' , '', l )
                            l = re.sub( r'Ga naar margenoot\+' , '', l )
                            full_text += l.strip() + '\n'
                        
                    
                    #print(full_text)
                    out.write(full_text)
                    
             
    out.close()
    print(f'Download saved as {title}.txt')
    
 
```

```python
download_texts( 'De chronyk van Antwerpen' , 'https://dbnl.nl/tekst/loon003antw01_01/')
download_texts( 'Chronycke van Antwerpen' , 'https://dbnl.nl/tekst/ulle001chro01_01/')
download_texts( 'Dagboek van Cornelis en Philip van Campene' , 'https://dbnl.nl/tekst/camp025dagb01_01/')
download_texts( 'Memoriaelbouc' , 'https://www.dbnl.nl/tekst/verw006memo01_01/')

```

```python
def download_single_text(title,url):

    print( f'Downloading {title} ...')
    title = re.sub( r'\s+' , '_' , title )
    out = open( f'{title}.txt' , 'w' , encoding = 'utf-8' )
    main_url = url

    response = requests.get(main_url)
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup( html ,"lxml")

        main_text = soup.find('table', {'id': 'tekst_en_noten'} )
        full_text = main_text.text.strip()
        lines = re.split( r'\n' , full_text)
        full_text = ''
        for i,l in enumerate(lines):
            if re.search( r'^voetnoot\d\)' , l ):
                break
            if re.search( r'^(margenoot\+)' , l ):
                l = ''
            if re.search( r'\w+' , l ):
                l = re.sub( r'\[pagina\s+\*?\w+\]' , '', l )
                l = re.sub( r'\[pagina\s+(\*)*\w+\]' , '', l )
                l = re.sub( r'\[p.\s+(\*)*\w+\]' , '', l )
                l = re.sub( r'Ga naar voetnoot\*' , '', l )
                l = re.sub( r'Ga naar margenoot\+' , '', l )
                l = re.sub( r'Ga naar eind\d+\)' , '', l )
                l = re.sub( r'Ga naar voetnoot\w+\)' , '', l )
                
                
                full_text += l.strip() + '\n'


        #print(full_text)
        out.write(full_text)

             
    out.close()
    print(f'Download saved as {title}.txt')
    
 
```

```python
download_single_text( 'Gerard Vremdt', 'https://www.dbnl.org/tekst/_bij019194101_01/_bij019194101_01_0017.php' )
download_single_text( 'Utrechtse kroniek over 1566-1576','https://www.dbnl.org/tekst/_bij005190401_01/_bij005190401_01_0006.php')
download_single_text( 'Kroniekje van Groningen' ,'https://www.dbnl.org/tekst/_bij005188901_01/_bij005188901_01_0004.php' )
```
